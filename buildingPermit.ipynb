#### indentify the numeric and categorical variables in the dataset

```python
#numeric variables
num_df = df.select_dtypes(include=['int64'])
num_df
```

```python
#categorical variables
cat_df = df.select_dtypes(exclude=['int64'])
cat_df
```

```python
cat_df1 = cat_df.columns.to_list()
cat_df1
```

```python
# Frequency table of each categorical variable
for col in cat_df1:
    val = df[col].value_counts()
    print(f"\t{col}:")
    print(f"{val} \n")
```
